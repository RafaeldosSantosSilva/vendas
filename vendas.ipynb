IMPORTAÇÕES


```python
import pandas as pd
import matplotlib.pyplot as plt
```

TRAZENDO A TABELA


```python
df = pd.read_csv('tabela_vendas.csv')
df
```

Manipulando a tabela

```python
df['Total das vendas'] = df['Preço Unitário'] * df['Quantidade']
df
```

```python
total_vendas_loja = df['Total das vendas'].sum()
print(f'O total vendido foi: {total_vendas_loja}')
```

Trazendo o total de cada produto vendido

```python
total_vendido_produto = df.groupby('Produto')['Total das vendas'].sum().sort_values(ascending=False)
total_vendido_produto
```

Trazendo o vendedor com maior numero de vendas

```python
total_vendas_vendedor = df.groupby('Vendedor')['Total das vendas'].sum().sort_values(ascending=False)
total_vendas_vendedor
```

```python
melhor_vendedor = total_vendas_vendedor.head(1)
print(f'''O melhor vendedor foi:
        {melhor_vendedor}''')
```

TRAZENDO GRAFICOS


```python
plt.figure(figsize=(10, 5))
plt.barh(total_vendido_produto.index, total_vendido_produto.values, color='skyblue')
plt.xlabel('Vendas')
plt.ylabel('Produtos')
plt.title('Ranking de Produtos Vendidos')
plt.gca().invert_yaxis()  
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.barh(total_vendas_vendedor.index, total_vendas_vendedor.values, color='navy')
plt.xlabel('Vendas')
plt.ylabel('Vendedores')
plt.title('Ranking de Vendedores')
plt.gca().invert_yaxis()
plt.show()
```
